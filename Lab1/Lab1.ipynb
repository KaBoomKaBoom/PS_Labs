```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Ts = 0.01
t = np.arange(0, 5, Ts)

# Generate white noise
x1 = np.random.rand(len(t))

# Plot the noise
plt.figure(figsize=(10, 6))
plt.plot(t, x1)
plt.title('1.1 White Noise')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(x1, bins=100, density=True)
plt.title('1.2 Histogram of White Noise')
plt.xlabel('Amplitude')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter

# Sampling parameters
Fs = 1000  # Sampling frequency (Hz)
Ts = 1 / Fs  # Sampling period
t = np.arange(0, 0.5, Ts)  # Time vector

# Generate white Gaussian noise
x1 = np.random.randn(len(t))

# Filter parameters
om0 = 2 * np.pi  # Natural frequency (rad/s)
dz = 0.005  # Damping factor
A = 1  # Gain
oms = om0 * Ts  # Normalized frequency

# Corrected Filter coefficients
b = [A * oms**2, 2 * A * oms**2, A * oms**2]  # Numerator coefficients
a = [1 + 2 * dz * oms + oms**2, -2 * (1 + dz * oms), 1]  # Denominator coefficients

# Normalize coefficients to make a[0] = 1
a = np.array(a) / a[0]
b = np.array(b) / a[0]

# Filter the noise
y1 = lfilter(b, a, x1)

# Plot the filtered signal
plt.figure(figsize=(10, 6))
plt.plot(t, y1)
plt.title('1.5 Filtered Noise with Second-Order Digital Filter')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
