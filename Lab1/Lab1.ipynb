```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter

# Define parameters
Ts = 0.001  # Sampling time
om0 = 2 * np.pi  # Natural frequency in rad/s
dz = 0.005  # Damping factor
A = 1  # Amplitude
oms = om0 * Ts  # Normalized frequency

# Filter coefficients
a = [1 + 2 * dz * oms + oms**2, -2 * (1 + dz * oms), 1]
b = [A * 2 * oms**2]

# Generate time vector
t = np.arange(0, 50, Ts)

# Generate white noise signal x1
x1 = np.random.rand(len(t))
plt.figure(figsize=(10, 5))
plt.plot(t, x1, label='Original signal')
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Original White Noise')
plt.legend()
plt.show()
# Apply the filter
y1 = lfilter(b, a, x1)

# Plot the filtered signal
plt.figure(figsize=(10, 5))
plt.plot(t, y1, label='Filtered Signal (y1)')
plt.grid(True)
plt.xlabel('Time (s)')
plt.ylabel('Function y(t)')
plt.title('Filtering Noise with a Second-Order Filter')
plt.legend()
plt.show()
```

```python
R = 50
m = np.arange(0, R)
s = 2 * m * (0.9**m)

plt.figure(figsize=(10, 6))
plt.stem(m, s)
plt.title('2.1 Original Signal')
plt.xlabel('Time Index n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
d = np.random.rand(len(m)) - 0.5

plt.figure(figsize=(10, 6))
plt.plot(m, s, label='Original Signal')
plt.plot(m, d, label='Random Signal')
plt.title('2.2/3 Original and Random Signals')
plt.xlabel('Time Index n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = s + d

plt.figure(figsize=(10, 6))
plt.plot(m, s, label='Original Signal')
plt.plot(m, x, label='Noisy Signal')
plt.title('2.4 Original and Noisy Signals')
plt.xlabel('Time Index n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def moving_average_filter(signal, M):
    b = np.ones(M) / M
    return lfilter(b, 1, signal)

# Apply MAF with M=3
M = 3
y = moving_average_filter(x, M)

plt.figure(figsize=(10, 6))
plt.plot(m, s, label='Original Signal')
plt.plot(m, x, label='Noisy Signal')
plt.plot(m, y, label='Filtered Signal (M=3)')
plt.title('2.5 Original, Noisy, and Filtered Signals')
plt.xlabel('Time Index n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
