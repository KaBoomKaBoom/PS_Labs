```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 500  # Number of samples
M = 32   # Filter order
mu = 0.01  # LMS step size
lambda_ = 0.99  # RLS forgetting factor

# Generate signals
x = np.random.randn(N)  # Input signal (white noise)
d = np.sin(0.1 * np.arange(N))  # Desired output (sine wave)

# LMS Algorithm
w_lms = np.zeros(M)
y_lms = np.zeros(N)
e_lms = np.zeros(N)

for n in range(M, N):
    x_n = x[n-M:n]
    y_lms[n] = np.dot(w_lms, x_n)
    e_lms[n] = d[n] - y_lms[n]
    w_lms = w_lms + mu * e_lms[n] * x_n

# RLS Algorithm
w_rls = np.zeros(M)
P = np.eye(M) * 1000  # Initializing the inverse correlation matrix
y_rls = np.zeros(N)
e_rls = np.zeros(N)

for n in range(M, N):
    x_n = x[n-M:n]
    phi = P @ x_n
    k = phi / (lambda_ + x_n.T @ phi)
    y_rls[n] = np.dot(w_rls, x_n)
    e_rls[n] = d[n] - y_rls[n]
    w_rls = w_rls + k * e_rls[n]
    P = (P - np.outer(k, phi)) / lambda_

# Plot Results
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(d, label='Desired signal')
plt.plot(y_lms, label='Filtered output (LMS)')
plt.plot(e_lms, label='Error signal (LMS)')
plt.title("LMS Adaptive Filtering")
plt.xlabel("Sample index")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(d, label='Desired signal')
plt.plot(y_rls, label='Filtered output (RLS)')
plt.plot(e_rls, label='Error signal (RLS)')
plt.title("RLS Adaptive Filtering")
plt.xlabel("Sample index")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Save results
np.save('adaptive_filtering_results.npy', {'LMS': {'weights': w_lms, 'error': e_lms, 'output': y_lms},
                                           'RLS': {'weights': w_rls, 'error': e_rls, 'output': y_rls}})

```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

class LMSFilter:
    def __init__(self, filter_length, mu):
        """
        Initialize LMS filter
        filter_length: Number of filter taps
        mu: Step size parameter
        """
        self.L = filter_length
        self.mu = mu
        self.w = np.zeros(self.L)
    
    def filter(self, x, d):
        """
        Apply LMS filtering
        x: Input signal
        d: Desired signal
        Returns: filtered signal, error signal
        """
        N = len(x)
        y = np.zeros(N)
        e = np.zeros(N)
        
        # Pad input for initial filter delay
        x_padded = np.pad(x, (self.L-1, 0), 'constant')
        
        for n in range(N):
            # Get current input window
            x_window = x_padded[n:n+self.L][::-1]
            
            # Calculate filter output
            y[n] = np.dot(self.w, x_window)
            
            # Calculate error
            e[n] = d[n] - y[n]
            
            # Update filter weights
            self.w = self.w + self.mu * e[n] * x_window
            
        return y, e

class RLSFilter:
    def __init__(self, filter_length, lambda_forget=0.99, delta=0.1):
        """
        Initialize RLS filter
        filter_length: Number of filter taps
        lambda_forget: Forgetting factor
        delta: Initial value for P matrix diagonal
        """
        self.L = filter_length
        self.lambda_forget = lambda_forget
        self.w = np.zeros(self.L)
        self.P = np.eye(self.L) / delta
    
    def filter(self, x, d):
        """
        Apply RLS filtering
        x: Input signal
        d: Desired signal
        Returns: filtered signal, error signal
        """
        N = len(x)
        y = np.zeros(N)
        e = np.zeros(N)
        
        # Pad input for initial filter delay
        x_padded = np.pad(x, (self.L-1, 0), 'constant')
        
        for n in range(N):
            # Get current input window
            x_window = x_padded[n:n+self.L][::-1]
            
            # Calculate Kalman gain
            k = self.P @ x_window / (self.lambda_forget + x_window.T @ self.P @ x_window)
            
            # Calculate filter output
            y[n] = np.dot(self.w, x_window)
            
            # Calculate error
            e[n] = d[n] - y[n]
            
            # Update filter weights
            self.w = self.w + k * e[n]
            
            # Update inverse correlation matrix
            self.P = (self.P - np.outer(k, x_window.T @ self.P)) / self.lambda_forget
            
        return y, e

# Example usage and demonstration
def demo_noise_cancellation():
    # Generate sample signals
    t = np.linspace(0, 1, 1000)
    desired_signal = np.sin(2 * np.pi * 10 * t)  # 10 Hz sine wave
    noise = 0.5 * np.random.randn(len(t))
    input_signal = desired_signal + noise
    
    # Apply LMS filter
    lms = LMSFilter(filter_length=32, mu=0.01)
    lms_output, lms_error = lms.filter(input_signal, desired_signal)
    
    # Apply RLS filter
    rls = RLSFilter(filter_length=32)
    rls_output, rls_error = rls.filter(input_signal, desired_signal)
    
    # Plot results
    plt.figure(figsize=(12, 8))
    
    plt.subplot(4, 1, 1)
    plt.plot(t, desired_signal)
    plt.title('Original Signal')
    plt.grid(True)
    
    plt.subplot(4, 1, 2)
    plt.plot(t, input_signal)
    plt.title('Noisy Signal')
    plt.grid(True)
    
    plt.subplot(4, 1, 3)
    plt.plot(t, lms_output)
    plt.title('LMS Filtered Signal')
    plt.grid(True)
    
    plt.subplot(4, 1, 4)
    plt.plot(t, rls_output)
    plt.title('RLS Filtered Signal')
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    demo_noise_cancellation()
```
