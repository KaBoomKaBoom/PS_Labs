```python
#Ex 2.1

import numpy as np
import matplotlib.pyplot as plt

# Time vector: 0 to 1 with step 1/99 (sampling frequency = 99 Hz)
t = np.arange(0, 1 + 1/99, 1/99)

# Generate the signal: sum of two sinusoids (15 Hz and 40 Hz)
x = np.sin(2 * np.pi * 15 * t) + np.sin(2 * np.pi * 40 * t)

# Compute the DFT using FFT
y = np.fft.fft(x)

# Compute magnitude and phase
m = np.abs(y)
p = np.unwrap(np.angle(y))  # Phase in radians

# Frequency vector: f = (0:length(y)-1) * 99 / length(y)
N = len(y)
f = np.arange(N) * 99 / N

# Plot magnitude spectrum
plt.figure()
plt.plot(f, m)
plt.title('Magnitude')
plt.xlabel('Frecvena (Hz)')
plt.ylabel('|Y(f)|')
plt.grid(True)
plt.gca().set_xticks([15, 40, 60, 85])
plt.gca().set_facecolor('white')
plt.gca().set_xlabel('Frecvena (Hz)', fontname='Arial', fontsize=12)
plt.gca().set_ylabel('|Y(f)|', fontname='Arial', fontsize=12)
plt.gca().set_title('Magnitude', fontname='Arial', fontsize=12)

# Plot phase spectrum (convert phase to degrees)
plt.figure()
plt.plot(f, p * 180 / np.pi)
plt.title('Faza')
plt.xlabel('Frecvena (Hz)')
plt.ylabel('Faza (grade)')
plt.grid(True)
plt.gca().set_xticks([15, 40, 60, 85])
plt.gca().set_facecolor('white')
plt.gca().set_xlabel('Frecvena (Hz)', fontname='Arial', fontsize=12)
plt.gca().set_ylabel('Faza (grade)', fontname='Arial', fontsize=12)
plt.gca().set_title('Faza', fontname='Arial', fontsize=12)

# Display the plots
plt.show()
```

```python
#Ex 4.3

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter

# Parameters
Ts = 0.01
T = 50

# Time vector
t = np.arange(0, T + Ts, Ts)

# Generate white noise signal
x1 = np.random.rand(len(t))

# Filter parameters
a = np.zeros(3)
b = np.zeros(1)
om0 = 2 * np.pi
dz = 0.05
A = 1
oms = om0 * Ts

# Compute filter coefficients
a[0] = 1 + 2 * dz * oms + oms**2
a[1] = -2 * (1 + dz * oms)
a[2] = 1
b[0] = A * 2 * dz * oms**2

# Apply the filter
y1 = lfilter(b, a, x1)

# Frequency parameters
df = 1 / T  # Frequency resolution
Fmax = 1 / Ts  # Maximum frequency
f = np.arange(-Fmax/2, Fmax/2 + df, df)

# Compute FFT for input and output signals
Fu1 = np.fft.fft(x1)
Fu2 = np.fft.fft(y1)

# Apply fftshift to center the spectra
Fu1p = np.fft.fftshift(Fu1)
Fu2p = np.fft.fftshift(Fu2)

# Compute magnitudes
m = np.abs(Fu1p)
m1 = np.abs(Fu2p)

# Ensure the length of f matches the length of the FFT results
f = f[:len(m)]

# Plot the magnitude spectrum of the input signal (x1)
plt.figure()
plt.stem(f, m, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.xlabel('Frecvena (Hz)')
plt.ylabel('|X1(f)|')
plt.title('Spectrul Modulului (Zgomot Alb - Intrare)')
plt.grid(True)
plt.gca().set_facecolor('white')
plt.gca().set_xlabel('Frecvena (Hz)', fontname='Arial', fontsize=12)
plt.gca().set_ylabel('|X1(f)|', fontname='Arial', fontsize=12)
plt.gca().set_title('Spectrul Modulului (Zgomot Alb - Intrare)', fontname='Arial', fontsize=12)

# Plot the magnitude spectrum of the output signal (y1)
plt.figure()
plt.stem(f, m1, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.xlabel('Frecvena (Hz)')
plt.ylabel('|Y1(f)|')
plt.title('Spectrul Modulului (Zgomot Alb - Ieire Filtrat)')
plt.grid(True)
plt.gca().set_facecolor('white')
plt.gca().set_xlabel('Frecvena (Hz)', fontname='Arial', fontsize=12)
plt.gca().set_ylabel('|Y1(f)|', fontname='Arial', fontsize=12)
plt.gca().set_title('Spectrul Modulului (Zgomot Alb - Ieire Filtrat)', fontname='Arial', fontsize=12)

# Display the plots
plt.show()
```

```python
#Ex. 5

import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 1000  # Sampling frequency (Hz)
T = 2  # Total time duration (seconds)
t = np.arange(-T/2, T/2, 1/fs)  # Time vector from -1 to 1 seconds

# Signal parameters
A = 1  # Amplitude
m = 0.5  # Modulation index
fc = 2  # Carrier frequency (Hz)
fm_values = [1, 2.5, 4]  # Modulation frequencies to test (Hz)

# Frequency vector for FFT
N = len(t)
df = fs / N  # Frequency resolution
f = np.arange(-fs/2, fs/2, df)  # Frequency vector from -fs/2 to fs/2

# Create subplots: one row per modulation frequency, two columns (time and frequency)
fig, axs = plt.subplots(len(fm_values), 2, figsize=(10, 3 * len(fm_values)))

for i, fm in enumerate(fm_values):
    # Generate the modulated signal
    s = A * (1 + m * np.cos(2 * np.pi * fm * t)) * np.cos(2 * np.pi * fc * t)

    # Compute the FFT and shift it
    S = np.fft.fft(s)
    S_shifted = np.fft.fftshift(S)
    mag = np.abs(S_shifted)

    # Plot the time-domain signal
    axs[i, 0].plot(t, s)
    axs[i, 0].set_xlabel('Time (Seconds)')
    axs[i, 0].set_ylabel('Waveform')
    axs[i, 0].set_title(f'Waveform (fm = {fm} Hz)')
    axs[i, 0].grid(True)
    axs[i, 0].set_xlim(-1, 1)
    axs[i, 0].set_facecolor('white')

    # Plot the magnitude spectrum (limit frequency axis to -6 to 6 Hz)
    axs[i, 1].plot(f, mag)
    axs[i, 1].set_xlabel('Frequency (Hertz)')
    axs[i, 1].set_ylabel('Magnitude')
    axs[i, 1].set_title(f'Magnitude (fm = {fm} Hz)')
    axs[i, 1].grid(True)
    axs[i, 1].set_xlim(-6, 6)
    axs[i, 1].set_facecolor('white')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

```python
#Ex.5 interface

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import tkinter as tk
from tkinter import ttk

# Signal parameters
fs = 1000  # Sampling frequency (Hz)
T = 2  # Total time duration (seconds)
t = np.arange(-T/2, T/2, 1/fs)  # Time vector from -1 to 1 seconds
A = 1  # Amplitude
m = 0.5  # Modulation index
fc = 2  # Carrier frequency (Hz)

# Frequency vector for FFT
N = len(t)
df = fs / N
f = np.arange(-fs/2, fs/2, df)

class CFTInterface:
    def __init__(self, root):
        self.root = root
        self.root.title("Continuous Fourier Transform Demo")

        # Create main frame
        self.main_frame = ttk.Frame(self.root, padding="10")
        self.main_frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

        # Create Matplotlib figure with two subplots
        self.fig, (self.ax1, self.ax2) = plt.subplots(1, 2, figsize=(10, 4))
        self.canvas = FigureCanvasTkAgg(self.fig, master=self.main_frame)
        self.canvas.get_tk_widget().grid(row=0, column=0, columnspan=3)

        # Initial modulation frequency
        self.fm = tk.DoubleVar(value=2.5)

        # Slider for modulation frequency
        ttk.Label(self.main_frame, text="Modulation Frequency (Hertz):").grid(row=1, column=0, pady=5)
        self.slider = ttk.Scale(self.main_frame, from_=0.5, to=5.0, orient=tk.HORIZONTAL, variable=self.fm, command=self.update_plots)
        self.slider.grid(row=1, column=1, pady=5)

        # Label to display current modulation frequency
        self.fm_label = ttk.Label(self.main_frame, text=f"{self.fm.get():.1f}")
        self.fm_label.grid(row=1, column=2, pady=5)

        # INFO button
        ttk.Button(self.main_frame, text="INFO", command=self.show_info).grid(row=2, column=0, pady=10)

        # Close button
        ttk.Button(self.main_frame, text="Close", command=self.root.quit).grid(row=2, column=2, pady=10)

        # Initial plot
        self.update_plots()

    def update_plots(self, *args):
        # Update the modulation frequency label
        self.fm_label.config(text=f"{self.fm.get():.1f}")

        # Clear previous plots
        self.ax1.clear()
        self.ax2.clear()

        # Generate the modulated signal
        fm_val = self.fm.get()
        s = A * (1 + m * np.cos(2 * np.pi * fm_val * t)) * np.cos(2 * np.pi * fc * t)

        # Compute the FFT and shift it
        S = np.fft.fft(s)
        S_shifted = np.fft.fftshift(S)
        mag = np.abs(S_shifted)

        # Plot the time-domain signal
        self.ax1.plot(t, s)
        self.ax1.set_xlabel('Time (Seconds)')
        self.ax1.set_ylabel('Waveform')
        self.ax1.set_title(f'Waveform (fm = {fm_val:.1f} Hz)')
        self.ax1.grid(True)
        self.ax1.set_xlim(-1, 1)

        # Plot the magnitude spectrum
        self.ax2.plot(f, mag)
        self.ax2.set_xlabel('Frequency (Hertz)')
        self.ax2.set_ylabel('Magnitude')
        self.ax2.set_title(f'Magnitude (fm = {fm_val:.1f} Hz)')
        self.ax2.grid(True)
        self.ax2.set_xlim(-6, 6)

        # Refresh the canvas
        self.canvas.draw()

    def show_info(self):
        # Create a new window for the INFO content
        info_window = tk.Toplevel(self.root)
        info_window.title("Continuous Fourier Transform Info")
        info_window.geometry("600x400")

        # Add a text widget to display the information
        info_text = tk.Text(info_window, wrap=tk.WORD, height=20, width=70)
        info_text.pack(pady=10, padx=10)

        # Information about the Continuous Fourier Transform
        info_content = """
        Continuous Fourier Transform (CFT) Information

        The Continuous Fourier Transform (CFT) is a mathematical tool used to transform a continuous-time signal x(t) from the time domain to the frequency domain. It is defined as:

        X(f) = _{-\infty}^{\infty} x(t) e^(-j 2 f t) dt

        where:
        - x(t) is the time-domain signal,
        - f is the frequency in Hertz,
        - X(f) is the frequency-domain representation of the signal,
        - j is the imaginary unit (j^2 = -1).

        Properties of the CFT:
        1. Linearity: F{a x(t) + b y(t)} = a X(f) + b Y(f)
        2. Time Shifting: F{x(t - t0)} = e^(-j 2 f t0) X(f)
        3. Frequency Shifting (Modulation): F{x(t) e^(j 2 f0 t)} = X(f - f0)
        4. Symmetry: If x(t) is real, then X(-f) = X*(f) (complex conjugate).

        Application in Modulation:
        For an amplitude-modulated signal s(t) = (1 + m cos(2 fm t)) cos(2 fc t), the CFT reveals:
        - A peak at the carrier frequency fc,
        - Two sidebands at fc  fm, with amplitudes proportional to the modulation index m.

        The CFT is fundamental in signal processing for analyzing the frequency content of signals, designing filters, and understanding modulation effects.
        """
        info_text.insert(tk.END, info_content)
        info_text.config(state='disabled')  # Make the text read-only

        # Add a close button for the info window
        ttk.Button(info_window, text="Close", command=info_window.destroy).pack(pady=5)

# Create the main application window
root = tk.Tk()
app = CFTInterface(root)
root.mainloop()
```
