```python
#Ex 1.2

import imageio.v3 as iio
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Read the image
# Replace 'cameraman.tif' with your image path
image_path = 'image.tiff'  # Update with your image path, e.g., 'C:/MATLAB/toolbox/images/imdemos/cameraman.tif'
c = iio.imread(image_path)

# Step 2: Crop a 256x256 region from the center
height, width = c.shape[:2]
start_y = (height - 256) // 2
start_x = (width - 256) // 2
cc = c[start_y:start_y + 256, start_x:start_x + 256] if c.ndim == 2 else c[start_y:start_y + 256, start_x:start_x + 256, :]

# Step 3: Add 128 to each pixel value
# Ensure pixel values stay within valid range (0-255 for uint8)
c1 = np.clip(cc.astype(np.int16) + 128, 0, 255).astype(np.uint8)

# Step 4: Display the original cropped and modified images
plt.figure(figsize=(10, 5))

# Original cropped image
plt.subplot(1, 2, 1)
plt.imshow(cc, cmap='gray' if cc.ndim == 2 else None)
plt.title('Original Cropped Image (Center)')
plt.axis('off')

# Modified image (after adding 128)
plt.subplot(1, 2, 2)
plt.imshow(c1, cmap='gray' if c1.ndim == 2 else None)
plt.title('Image with 128 Added to Pixels')
plt.axis('off')

plt.tight_layout()
plt.show()
```
